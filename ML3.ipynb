```
import matplotlib.pyplot as plt
import numpy as np

def Gaussian1D(x, mu, sigma):
  num = np.exp(-0.5*((x-mu)/sigma)**2)
  den = np.sqrt(2*np.pi*sigma)
  return(num/den)
  
mu,sigma = 1, 10
x = np.linspace(-5,5)
y = Gaussian1D(x, mu, sigma)
plt.plot(x,y)
plt.show()
```

```
from scipy.stats import multivariate_normal
x =np.linspace(-5,5,100)
y = multivariate_normal.pdf(x,mean=1,cov=1)
plt.plot(x,y)
plt.show()
```

**N-D Gaussian Scipy**

```
from matplotlib import cm

x, y = np.mgrid[-5:5:.01, -5:5:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
rv = multivariate_normal(mean = [1, 1], cov = [[1, 0], [0, 1]])

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(121, projection='3d')
plt.xlabel('x')
plt.ylabel('y')
ax.plot_surface(x,y,rv.pdf(pos), cmap=cm.Blues)
plt.subplot(122)
plt.contourf(x, y, rv.pdf(pos), cmap=cm.Blues)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```
